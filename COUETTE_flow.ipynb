# Couette flow solution

Let $\eta=\dfrac{y}{h}$ such that $0 \le \eta \le 1$.

The velocity profile is given by

$$\frac{u}{U}=\dfrac{h^2}{2\mu U}\left(\dfrac{\text{d}p^\prime}{\text{d}x}\right)\left(\eta^2-\eta\right)+\eta$$


# Plot as a function of a reference pressure gradient

Evaluating $\left.\dfrac{\text{d}(u/U)}{\text{d}\eta}\right|_{\eta=0}=0$, implies a pressure gradient of

$$\dfrac{\text{d}p{'}}{\text{d}x}=\dfrac{2\mu U}{h^2}$$

So let's plot the velocity profile for as a function of a constant $\Lambda$ such that

$$\dfrac{\text{d}p{'}}{\text{d}x}=\dfrac{2\mu U}{h^2} \Lambda$$

giving


$$\frac{u}{U}=\Lambda\left(\eta^2-\eta\right)+\eta=\left(\Lambda\left(\eta-1\right)+1\right)\eta$$

and

$$\dfrac{\text{d}}{\text{d}\eta}\left(\frac{u}{U}\right)=\Lambda\left(2\eta-1\right)+1$$


```python
fig, (ax1, ax2) = mpl.subplots(1,2)
fig.subplots_adjust( wspace = .1 )

Lambda_lst = ( 2, 1, 0.5, 0.0, -0.5, -1.0, -2.0, -3.3 )

for Lambda in Lambda_lst:
  eta = np.linspace( 0, 1, 100 )
  vel = (Lambda * (eta-1.0)+1)*eta
  dvel = Lambda * (2*eta-1.0)+1

  if Lambda == 1:
    ax1.plot( vel, eta, dashes=(8,2,2,2), label="$\Lambda=%+.1f$" % Lambda )
  elif Lambda == 0:
    ax1.plot( vel, eta, lw=2.5,dashes=(8,2), label="$\Lambda=%.1f$" % Lambda )
  else:
    ax1.plot( vel, eta, label="$\Lambda=%+.1f$" % Lambda )

  if Lambda == 1:
    ax2.plot( dvel, eta, dashes=(8,2,2,2), label="$\Lambda=%+.1f$" % Lambda )
  elif Lambda == 0:
    ax2.plot( dvel, eta, lw=2.5,dashes=(8,2), label="$\Lambda=%.1f$" % Lambda )
  else:
    ax2.plot( dvel, eta, label="$\Lambda=%+.1f$" % Lambda )


ax1.set_ylabel( "$\eta$" )
ax1.set_xlabel( "$u/U$" )
ax1.legend()
ax1.set_yticks(np.arange( 0, 1.01, 1) )
ax1.set_xticks(np.arange( 0, 1.01, 1) )
ax1.grid()
ax1.
ax2.set_ylabel( "$\eta$" )
ax2.set_xlabel( "$u/U$" )
ax2.legend()
ax2.set_yticks(np.arange( 0, 1.01, 1) )
ax2.set_xticks(np.arange( 0, 1.01, 1) )
ax2.grid();
```
