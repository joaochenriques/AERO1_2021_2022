# Couette flow solution

Let $\eta=\dfrac{y}{h}$ such that $0 \le \eta \le 1$.

The velocity profile is given by

$$\frac{u}{U}=\dfrac{h^2}{2\mu U}\left(\dfrac{\text{d}p^\prime}{\text{d}x}\right)\left(\eta^2-\eta\right)+\eta$$


# Plot as a function of a reference pressure gradient

The $\left.\dfrac{\text{d}(u/U)}{\text{d}\eta}\right|_{\eta=0}=0$ for a pressure gradient of

$$\dfrac{\text{d}p^\prime}{\text{d}x}=\dfrac{2\mu U}{h^2}$$

So let's plot the velocity profile for as a function of a constant $\Lambda$

$$\dfrac{\text{d}p^\prime}{\text{d}x}=\dfrac{2\mu U}{h^2} \Lambda$$

giving


$$\frac{u}{U}=\Lambda\left(\eta^2-\eta\right)+\eta=\left(\Lambda\left(\eta-1\right)+1\right)\eta$$


```python
Lambda_lst = ( 2, 1, 0.5, 0.0, -0.5, -1.0, -2.0, -3.0 )

for Lambda in Lambda_lst:
  eta = np.linspace( 0, 1, 100 )
  vel = (Lambda * (eta-1.0)+1)*eta

  if Lambda == 1:
    mpl.plot( vel, eta, dashes=(8,2,2,2), label="$\Lambda=%+.1f$" % Lambda )
  elif Lambda == 0:
    mpl.plot( vel, eta, lw=2.5,dashes=(8,2), label="$\Lambda=%.1f$" % Lambda )
  else:
    mpl.plot( vel, eta, label="$\Lambda=%+.1f$" % Lambda )

mpl.ylabel( "$\eta$" )
mpl.xlabel( "$u/U$" )
mpl.legend()
mpl.gca().set_yticks(np.arange( 0, 1.01, 1) )
mpl.gca().set_xticks(np.arange( 0, 1.01, 1) )
mpl.grid();
```
